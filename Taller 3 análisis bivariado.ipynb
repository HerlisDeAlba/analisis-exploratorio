### Diccionario de datos

# IV: Relationships between variables

```python
sns.pairplot(df)
```

Visualmente se observa un comportamiento lineal entre promedio de horas trabajadas al mes y la antigüedad

```python
# Analizamos la relación entre promedio de horas trabajadas al mes y los que dejaron la empresa

sns.catplot(x="left", y="prom_hrs_mes", data=df, kind="box", aspect=1.5)
plt.title(" Promedio de hrs trabajadas al mes vs Left")
plt.show()
```

Los que dejaron la empresa trabajaban al mes, un poco más que los que continuan en la compañía.

```python
# Analizamos visualmente la relación entre antigüedad y los que dejaron la empresa

sns.catplot(x="left", y="antigüedad", data=df, kind="box", aspect=1.5)
plt.title("Boxplot antigüedad vs y los que dejaron la empresa")
plt.show()
```

Quienes se fueron, tenían entre 3 y 9 años en la compañía, pero la mayoría, estaban alrededor de los 7 años de servicio.

```python
# Relación entre promedio de horas trabajadas al mes, antigüedad y los que dejaron la empresa

sns.scatterplot(x="antigüedad", y="prom_hrs_mes", hue="left", data=df, palette="Dark2", s=80)
plt.title("antigüedad, prom_hrs_mes y left")
plt.show()
```

En su mayoría, hay 2 grupos de empleados que tenderán a dejar la empresa:

1. Los empleados que tienen a lo máximo 3 años de servicio y trabajan menos de 175 horas al mes.
2. Los empleados que tienen entre 7 y 8 años de servicio y laboran entre 184 y 190 horas al mes.

## Análisis

Poco menos del 30% de los empleados han dejado la compañía. Al analizar la gráfica que relaciona varias variables, se observa una conexión particular entre las horas trabajadas al mes y la antigüedad. Al enfocar estos factores en los empleados que dejaron la empresa, se puede concluir que aquellos que se marcharon trabajaban, en promedio, un poco más que los que permanecen en la compañía. Además, quienes dejaron la empresa tenían entre 3 y 9 años de antigüedad, con la mayoría alrededor de los 7 años de servicio.
Al observar el comportamiento de las tres variables (horas trabajadas al mes, antigüedad y estado de permanencia en la empresa), se infiere que hay dos grupos de empleados con mayor probabilidad de dejar la empresa:
1.	Empleados con un máximo de 3 años de servicio que trabajan menos de 175 horas al mes.
2.	Empleados con entre 7 y 8 años de servicio que laboran entre 184 y 190 horas al mes.
